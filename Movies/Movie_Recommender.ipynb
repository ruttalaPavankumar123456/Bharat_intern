## Feature Selection

```python
movies.columns
```

```python
movie = movies[['id','title','genre','overview']]
```

```python
movie
```

```python
movie['tags'] = movie['genre']+movie['overview']
```

```python
movie
```

```python
drop_data = movie.drop(columns= 'genre')
```

```python
drop_data = drop_data.drop(columns= 'overview')

# we can also use to combinely removal of overview and genre
#    drop_data = movie.drop(columns=['genre', 'overview'])
```

```python
drop_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=10000, stop_words= 'english')
```

```python
cv
```

```python
vector = cv.fit_transform(drop_data['tags'].values.astype('U')).toarray()
```

```python
vector
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
drop_data[drop_data['title']=="The Godfather"]
```

```python
drop_data[drop_data['title']=="The Godfather"].index[0]
```

```python
drop_data.info()
```

```python
distance = sorted(list(enumerate(similarity[2])),  reverse=True, key = lambda vector:vector[1])
for i in distance[0:5]:
    print(drop_data.iloc[i[0]].title)
```

```python
def recommand(movie):
    index = drop_data[drop_data['title']==movie].index[0]
    distance = sorted(list(enumerate(similarity[index])),  reverse=True, key = lambda vector:vector[1])
    for i in distance[0:5]:
        print(drop_data.iloc[i[0]].title)
```

```python
recommand('The Shawshank Redemption')
```
